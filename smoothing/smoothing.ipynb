```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
```

```python
# open pickle file 
with open('../results/25days/predictions_0_1.pkl', 'rb') as f:
    df_dict = pickle.load(f)

rolled_df = pd.DataFrame()
for key, value in df_dict.items():
    #if key is not 'Real Temperature' or 'IEEE 738'
    
    current_df = df_dict[key]
    current_df_melted = pd.melt(current_df, id_vars=['datetime'], value_vars=current_df.columns[1:], var_name='measurement', value_name='value')
    print(current_df_melted.isna().sum())
    current_df_melted_sorted = current_df_melted.sort_values(['datetime', 'measurement'])
    current_df_value = current_df_melted_sorted['value']
    #reset index
    current_df_value = current_df_value.reset_index(drop=True)
    if (key != 'Real Temperature') and (key != 'IEEE 738'):
        current_df_value_rolled = current_df_value.rolling(window=15, min_periods=1).mean()
        rolled_df[key] = current_df_value_rolled
    else:
        rolled_df[key] = current_df_value
```
